```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('creditcard copy.csv')
pd.options.display.max_columns = 100
```

```python
df.head(10)
```

```python
df.value_counts('Class')[1] / df.shape[0] * 100
```

Объектов класса 1 всего **0.17%** => эта выборка несбалансированная

```python
df.info()
```

Да, все объекты имеют числовой тип данных и не имеют пропусков.

```python
x = df.drop('Class', axis = 1)
```

```python
y = pd.Series(df['Class'])
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=100)
```

```python
parameters = [{'n_estimators': [10, 15],
               'max_features': np.arange(3, 5),
               'max_depth': np.arange(4, 7)}]
```

```python
clf = GridSearchCV(estimator=RandomForestClassifier(random_state=100),
                   param_grid=parameters,
                   scoring='roc_auc',
                   cv=3)
```

```python
clf.fit(x_train, y_train)
```

```python
clf.best_params_
```

```python
y_pred = clf.predict(x_test)
```

```python
y_proba = clf.predict_proba(x_test)
y_proba
```

```python
y_proba.shape[0]
```

```python
y_proba_mass = np.array([])
for n in range(y_proba.shape[0]):
    print(y_proba[n][1])
    y_proba_mass = np.append(y_proba_mass, y_proba[n][1])
```

```python
y_proba_mass[0:3]
```

```python
from sklearn.metrics import roc_auc_score
```

```python
roc_auc_score(y_test, y_proba_mass)
```
